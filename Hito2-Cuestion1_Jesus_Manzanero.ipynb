### ¿Que es un DataLake? Un DataLake es un repositorio de almacenamiento que contienen una gran cantidad de datos en bruto y que se mantienen allí hasta que sea necesario

```python
import pandas as pd
datosjson = pd.read_json(r"https://raw.githubusercontent.com/Chexu28/Hito2_Jesus_Manzanero/main/us-states.json")
datosjson.head(10)
```

### El método describe muestra devuelve información estadística de los datos del dataframe o de la serie.                Esta información incluye el número de muestras, el valor medio, la desviación estándar, el valor mínimo, máximo, la media y los valores correspondientes a los porcentajes 25% y 75%.

```python
datosjson.describe()
```

### El metodo info muestra un resumen de un dataframe, incluyendo información sobre el tipo de los índices de filas y columnas, los valores no nulos y la memoria usada
