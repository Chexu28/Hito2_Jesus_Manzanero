```python
import pandas as pd
data = pd.read_csv(r"https://raw.githubusercontent.com/Chexu28/Hito2_Jesus_Manzanero/main/Worldwide%20Vaccine%20Data.csv")
data.head(10)
```

### Gráfico circular con los porcentajes de los 10 paises más vacunados

```python
import matplotlib.pyplot as plt
dataordenado = data.sort_values('% of population vaccinated', ascending=False)
datafinal = dataordenado.head(10)
#print(datafinal)
datapais = datafinal['Country']
dataporcentajes = datafinal['% of population vaccinated']
plt.pie(dataporcentajes, labels = datapais, autopct='%1.1f%%')

plt.show()
```

### Gráfico de barras donde se muestran las dosis administradas por los 5 primeros paises

```python

datatotal = datafinal['Total doses administered'].head()
datapais = datafinal['Country'].head()
plt.bar(datapais, datatotal)
plt.show()
```

```python
from seaborn import distplot


distplot(dataporcentajes)

```

```python
import numpy as np
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
fig = plt.figure(figsize = (9, 5))
ax = fig.gca(projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap = "summer")
fig.colorbar(surface)
plt.show()
```
